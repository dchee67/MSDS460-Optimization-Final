```python
# Wiring dictionaries

products = ["Widgets", "Gadgets", "Flugels"]
periods = [1, 2, 3, 4, 5]
plants = ["A", "B"]

# Contract demand by product/period
demand = {
    "Widgets": {1: 70,  2: 125, 3: 185, 4: 190, 5: 200},
    "Gadgets": {1: 200, 2: 300, 3: 295, 4: 245, 5: 240},
    "Flugels": {1: 140, 2: 175, 3: 205, 4: 235, 5: 230},
}

# Advertising "unit demand created" variables (ads in t create extra demand in t+1)
ads = {
    "Widgets": {1: widgets_advert_p1, 2: widgets_advert_p2, 3: widgets_advert_p3, 4: widgets_advert_p4},
    "Gadgets": {1: gadgets_advert_p1, 2: gadgets_advert_p2, 3: gadgets_advert_p3, 4: gadgets_advert_p4},
    "Flugels": {1: flugels_advert_p1, 2: flugels_advert_p2, 3: flugels_advert_p3, 4: flugels_advert_p4},
}

# Cost per 1 additional unit of demand (ad spend per "advert unit")
ad_cost_per_unit = {"Widgets": 160, "Gadgets": 120, "Flugels": 180}

# Production variables mapping
prod = {
    "A": {
        "Widgets": {1: widgets_produced_p1_A, 2: widgets_produced_p2_A, 3: widgets_produced_p3_A, 4: widgets_produced_p4_A, 5: widgets_produced_p5_A},
        "Gadgets": {1: gadgets_produced_p1_A, 2: gadgets_produced_p2_A, 3: gadgets_produced_p3_A, 4: gadgets_produced_p4_A, 5: gadgets_produced_p5_A},
        "Flugels": {1: flugels_produced_p1_A, 2: flugels_produced_p2_A, 3: flugels_produced_p3_A, 4: flugels_produced_p4_A, 5: flugels_produced_p5_A},
    },
    "B": {
        "Widgets": {1: widgets_produced_p1_B, 2: widgets_produced_p2_B, 3: widgets_produced_p3_B, 4: widgets_produced_p4_B, 5: widgets_produced_p5_B},
        "Gadgets": {1: gadgets_produced_p1_B, 2: gadgets_produced_p2_B, 3: gadgets_produced_p3_B, 4: gadgets_produced_p4_B, 5: gadgets_produced_p5_B},
        "Flugels": {1: flugels_produced_p1_B, 2: flugels_produced_p2_B, 3: flugels_produced_p3_B, 4: flugels_produced_p4_B, 5: flugels_produced_p5_B},
    },
}

# Inventory "stored after period t" exists for t=1..4 (end inv after period 5 must be 0)
inv = {
    "A": {
        "Widgets": {1: widgets_stored_p1_A, 2: widgets_stored_p2_A, 3: widgets_stored_p3_A, 4: widgets_stored_p4_A},
        "Gadgets": {1: gadgets_stored_p1_A, 2: gadgets_stored_p2_A, 3: gadgets_stored_p3_A, 4: gadgets_stored_p4_A},
        "Flugels": {1: flugels_stored_p1_A, 2: flugels_stored_p2_A, 3: flugels_stored_p3_A, 4: flugels_stored_p4_A},
    },
    "B": {
        "Widgets": {1: widgets_stored_p1_B, 2: widgets_stored_p2_B, 3: widgets_stored_p3_B, 4: widgets_stored_p4_B},
        "Gadgets": {1: gadgets_stored_p1_B, 2: gadgets_stored_p2_B, 3: gadgets_stored_p3_B, 4: gadgets_stored_p4_B},
        "Flugels": {1: flugels_stored_p1_B, 2: flugels_stored_p2_B, 3: flugels_stored_p3_B, 4: flugels_stored_p4_B},
    },
}

# Labor constraints (per plant, per period)

plants = ["A", "B"]
products = ["Widgets", "Gadgets", "Flugels"]
periods = [1, 2, 3, 4, 5]

# Labor hours required per unit produced
labor_use = {
    "A": {"Widgets": 9.5, "Gadgets": 7.1, "Flugels": 11.1},
    "B": {"Widgets": 9.1, "Gadgets": 7.8, "Flugels": 10.6},
}

# Regular labor capacity per period
reg_cap = {"A": 2500, "B": 3800}

# New labor decision vars (keep your existing labor vars untouched)
reg_hours = lp.LpVariable.dicts("RegHours", (plants, periods), lowBound=0)
ot_hours  = lp.LpVariable.dicts("OtHours",  (plants, periods), lowBound=0)

# Plant storage capacities
storage_cap = {"A": 70, "B": 50}
```

```python
# Extra demand fulfilled (units) due to advertising
# extra sales in periods 2-5 cannot exceed the advertising-generated demand from the prior period
extra_sales = lp.LpVariable.dicts(
    "ExtraSales",
    (products, [2, 3, 4, 5]),
    lowBound=0
)
```

```python
# Advertising creates extra demand next period (cap extra sales by prior period ads)
for i in products:
    for t in [2, 3, 4, 5]:
        model += extra_sales[i][t] <= ads[i][t-1], f"AdLink_{i}_t{t}"

# Total advertising budget (spend = cost_per_unit * advert_units)
model += lp.lpSum(ad_cost_per_unit[i] * ads[i][t] for i in products for t in [1,2,3,4]) <= 70000, "AdBudget"
```

```python
# Helper functions for clean expressions
def total_prod(i, t):
    return prod["A"][i][t] + prod["B"][i][t]

def total_inv_end(i, t):  # t=1..4 only
    return inv["A"][i][t] + inv["B"][i][t]

def total_inv_prev(i, t):  # previous end inventory for t>=2
    return total_inv_end(i, t-1)

# Period 1 balance: Production = contract demand + ending inventory after period 1
for i in products:
    model += total_prod(i, 1) == demand[i][1] + total_inv_end(i, 1), f"InventoryFlow_{i}_t1"

# Periods 2-4: prior inv + production = contract demand + extra sales + ending inv
for i in products:
    for t in [2, 3, 4]:
        model += total_inv_prev(i, t) + total_prod(i, t) == demand[i][t] + extra_sales[i][t] + total_inv_end(i, t), f"InventoryFlow_{i}_t{t}"

# Period 5: prior inv + production = contract demand + extra sales, and ending inventory forced to 0
for i in products:
    model += total_inv_prev(i, 5) + total_prod(i, 5) == demand[i][5] + extra_sales[i][5], f"InventoryFlow_{i}_t5"
```

The objective problem will be maximizing the profit of the company, which is : $ from sales - ( advertising costs + storage costs + material costs + transportation costs + labor costs ).

```python
model += lp.lpSum(
    2490 * lp.lpSum(widgets_produced_p1_A + widgets_produced_p2_A + widgets_produced_p3_A + widgets_produced_p4_A + widgets_produced_p5_A + widgets_produced_p1_B + widgets_produced_p2_B + widgets_produced_p3_B + widgets_produced_p4_B + widgets_produced_p5_B)
    + 1990 * lp.lpSum(gadgets_produced_p1_A + gadgets_produced_p2_A + gadgets_produced_p3_A + gadgets_produced_p4_A + gadgets_produced_p5_A + gadgets_produced_p1_B + gadgets_produced_p2_B + gadgets_produced_p3_B + gadgets_produced_p4_B + gadgets_produced_p5_B)
    + 2970 * lp.lpSum(flugels_produced_p1_A + flugels_produced_p2_A + flugels_produced_p3_A + flugels_produced_p4_A + flugels_produced_p5_A + flugels_produced_p1_B + flugels_produced_p2_B + flugels_produced_p3_B + flugels_produced_p4_B + flugels_produced_p5_B)

    - 160 * lp.lpSum(widgets_advert_p1 + widgets_advert_p2 + widgets_advert_p3 + widgets_advert_p4)
    - 120 * lp.lpSum(gadgets_advert_p1 + gadgets_advert_p2 + gadgets_advert_p3 + gadgets_advert_p4)
    - 180 * lp.lpSum(flugels_advert_p1 + flugels_advert_p2 + flugels_advert_p3 + flugels_advert_p4)

    - 7.5 * lp.lpSum(widgets_stored_p1_A + widgets_stored_p2_A + widgets_stored_p3_A + widgets_stored_p4_A)
    - 5.5 * lp.lpSum(gadgets_stored_p1_A + gadgets_stored_p2_A + gadgets_stored_p3_A + gadgets_stored_p4_A)
    - 6.5 * lp.lpSum(flugels_stored_p1_A + flugels_stored_p2_A + flugels_stored_p3_A + flugels_stored_p4_A)
    - 7.8 * lp.lpSum(widgets_stored_p1_B + widgets_stored_p2_B + widgets_stored_p3_B + widgets_stored_p4_B)
    - 5.7 * lp.lpSum(gadgets_stored_p1_B + gadgets_stored_p2_B + gadgets_stored_p3_B + gadgets_stored_p4_B)
    - 7.0 * lp.lpSum(flugels_stored_p1_B + flugels_stored_p2_B + flugels_stored_p3_B + flugels_stored_p4_B)

    - 1.25 * lp.lpSum(RM1_A_p1 + RM1_A_p2 + RM1_A_p3 + RM1_A_p4 + RM1_A_p5)
    - 2.65 * lp.lpSum(RM2_A_p1 + RM2_A_p2 + RM2_A_p3 + RM2_A_p4 + RM2_A_p5)
    - 1.45 * lp.lpSum(RM1_B_p1 + RM1_B_p2 + RM1_B_p3 + RM1_B_p4 + RM1_B_p5)
    - 2.90 * lp.lpSum(RM2_B_p1 + RM2_B_p2 + RM2_B_p3 + RM2_B_p4 + RM2_B_p5)

    - 6.30 * A_widget_trans
    - 4.60 * A_gadget_trans
    - 5.50 * A_flugel_trans
    - 6.50 * B_widget_trans
    - 5.00 * B_gadget_trans
    - 5.70 * B_flugel_trans

    - lp.lpSum(11 * reg_hours["A"][t] + 16.5 * ot_hours["A"][t] for t in [1, 2])
    - lp.lpSum(11 * reg_hours["B"][t] + 16.5 * ot_hours["B"][t] for t in [1, 2])
    - lp.lpSum((11 * 1.05) * reg_hours["A"][t] + (16.5 * 1.05) * ot_hours["A"][t] for t in [3, 4, 5])
    - lp.lpSum((11 * 1.10) * reg_hours["B"][t] + (16.5 * 1.10) * ot_hours["B"][t] for t in [3, 4, 5])
)
```

## Constraints

```python
# Regular time caps
for p in plants:
    for t in periods:
        model += reg_hours[p][t] <= reg_cap[p], f"RegCap_{p}_t{t}"

# Labor balance: production labor requirement = regular + overtime hours
for p in plants:
    for t in periods:
        model += (
            lp.lpSum(labor_use[p][i] * prod[p][i][t] for i in products)
            == reg_hours[p][t] + ot_hours[p][t]
        ), f"LaborBalance_{p}_t{t}"
```

```python
# Raw material availability constraints (vendor caps per period)

# Raw material usage (lbs) per unit produced
rm1_use = {
    "A": {"Widgets": 194, "Gadgets": 230, "Flugels": 178},
    "B": {"Widgets": 188, "Gadgets": 225, "Flugels": 170},
}
rm2_use = {
    "A": {"Widgets": 8.6, "Gadgets": 0.0, "Flugels": 11.6},
    "B": {"Widgets": 9.2, "Gadgets": 0.0, "Flugels": 10.8},
}

# RM1_A, RM2_A = total raw material used at Plant A
# RM1_B, RM2_B = total raw material used at Plant B
# RM1, RM2     = total raw material used across both plants

periods_RM = {
    1: [RM1_A_p1, RM2_A_p1, RM1_B_p1, RM2_B_p1],
    2: [RM1_A_p2, RM2_A_p2, RM1_B_p2, RM2_B_p2],
    3: [RM1_A_p3, RM2_A_p3, RM1_B_p3, RM2_B_p3],
    4: [RM1_A_p4, RM2_A_p4, RM1_B_p4, RM2_B_p4],
    5: [RM1_A_p5, RM2_A_p5, RM1_B_p5, RM2_B_p5]
}


for t in periods:
    model += (
        periods_RM[t][0] ==
        lp.lpSum(rm1_use["A"][i] * prod["A"][i][t] for i in products)
    ), f"RM1_A_Accounting_p{t}"

    model += (
        periods_RM[t][1] ==
        lp.lpSum(rm2_use["A"][i] * prod["A"][i][t] for i in products)
    ), f"RM@_A_Accounting_p{t}"

    # Plant B total raw material usage
    model += (
        periods_RM[t][2] ==
        lp.lpSum(rm1_use["B"][i] * prod["B"][i][t] for i in products)
    ), f"RM1_B_Accounting_p{t}"

    model += (
        periods_RM[t][3] ==
        lp.lpSum(rm2_use["B"][i] * prod["B"][i][t] for i in products)
    ), f"RM2_B_Accounting_p{t}"

    model += (
        periods_RM[t][0] + periods_RM[t][2] <= 70 * 2000
    ), f"RM1_Accounting_p{t}"

    model += (
        periods_RM[t][1] + periods_RM[t][3] <= 2.5 * 2000
    ), f"RM2_Accounting_p{t}"
```

```python
# TRANSPORTATION BALANCE CONSTRAINTS
# total shipments from plant p of product i = total production at plant p of product i across all periods


# Widgets shipped from Plant A must equal total Widgets produced at Plant A
model += (
    A_widget_trans ==
    widgets_produced_p1_A + widgets_produced_p2_A + widgets_produced_p3_A +
    widgets_produced_p4_A + widgets_produced_p5_A
), "TransportBalance_A_Widgets"

# Gadgets shipped from Plant A must equal total Gadgets produced at Plant A
model += (
    A_gadget_trans ==
    gadgets_produced_p1_A + gadgets_produced_p2_A + gadgets_produced_p3_A +
    gadgets_produced_p4_A + gadgets_produced_p5_A
), "TransportBalance_A_Gadgets"

# Flugels shipped from Plant A must equal total Flugels produced at Plant A
model += (
    A_flugel_trans ==
    flugels_produced_p1_A + flugels_produced_p2_A + flugels_produced_p3_A +
    flugels_produced_p4_A + flugels_produced_p5_A
), "TransportBalance_A_Flugels"

# Widgets shipped from Plant B must equal total Widgets produced at Plant B
model += (
    B_widget_trans ==
    widgets_produced_p1_B + widgets_produced_p2_B + widgets_produced_p3_B +
    widgets_produced_p4_B + widgets_produced_p5_B
), "TransportBalance_B_Widgets"

# Gadgets shipped from Plant B must equal total Gadgets produced at Plant B
model += (
    B_gadget_trans ==
    gadgets_produced_p1_B + gadgets_produced_p2_B + gadgets_produced_p3_B +
    gadgets_produced_p4_B + gadgets_produced_p5_B
), "TransportBalance_B_Gadgets"

# Flugels shipped from Plant B must equal total Flugels produced at Plant B
model += (
    B_flugel_trans ==
    flugels_produced_p1_B + flugels_produced_p2_B + flugels_produced_p3_B +
    flugels_produced_p4_B + flugels_produced_p5_B
), "TransportBalance_B_Flugels"


```

```python
# STORAGE CAPACITY CONSTRAINTS
# the total number of finished goods stored at the end of each period does not exceed each plant's available storage capacity
# Plant A can store at most 70 total units across all products
# Plant B can store at most 50 total units across all products
# inventory is only tracked after periods 1-4

storage_by_period = {
    1: {
        "A": {"Widgets": widgets_stored_p1_A, "Gadgets": gadgets_stored_p1_A, "Flugels": flugels_stored_p1_A},
        "B": {"Widgets": widgets_stored_p1_B, "Gadgets": gadgets_stored_p1_B, "Flugels": flugels_stored_p1_B},
    },
    2: {
        "A": {"Widgets": widgets_stored_p2_A, "Gadgets": gadgets_stored_p2_A, "Flugels": flugels_stored_p2_A},
        "B": {"Widgets": widgets_stored_p2_B, "Gadgets": gadgets_stored_p2_B, "Flugels": flugels_stored_p2_B},
    },
    3: {
        "A": {"Widgets": widgets_stored_p3_A, "Gadgets": gadgets_stored_p3_A, "Flugels": flugels_stored_p3_A},
        "B": {"Widgets": widgets_stored_p3_B, "Gadgets": gadgets_stored_p3_B, "Flugels": flugels_stored_p3_B},
    },
    4: {
        "A": {"Widgets": widgets_stored_p4_A, "Gadgets": gadgets_stored_p4_A, "Flugels": flugels_stored_p4_A},
        "B": {"Widgets": widgets_stored_p4_B, "Gadgets": gadgets_stored_p4_B, "Flugels": flugels_stored_p4_B},
    },
}


for t in [1, 2, 3, 4]:
    for p in plants:
        model += (
            lp.lpSum(storage_by_period[t][p][i] for i in products)
            <= storage_cap[p]
        ), f"StorageCap_{p}_t{t}"
```

```python
# Baseline storage capacities
BASE_STORAGE_A = 70
BASE_STORAGE_B = 50

# Storage scenarios
storage_scenarios = [
    {"Scenario": "Baseline",        "A_mult": 1.00, "B_mult": 1.00},
    {"Scenario": "Plant A +10%",    "A_mult": 1.10, "B_mult": 1.00},
    {"Scenario": "Plant A +20%",    "A_mult": 1.20, "B_mult": 1.00},
    {"Scenario": "Plant A +30%",    "A_mult": 1.30, "B_mult": 1.00},
    {"Scenario": "Plant B +10%",    "A_mult": 1.00, "B_mult": 1.10},
    {"Scenario": "Plant B +20%",    "A_mult": 1.00, "B_mult": 1.20},
    {"Scenario": "Plant B +30%",    "A_mult": 1.00, "B_mult": 1.30},
    {"Scenario": "Both +10%",       "A_mult": 1.10, "B_mult": 1.10},
    {"Scenario": "Both +20%",       "A_mult": 1.20, "B_mult": 1.20},
    {"Scenario": "Both +30%",       "A_mult": 1.30, "B_mult": 1.30},
]

# Helper metrics
def total_overtime():
    return sum(lp.value(ot_hours[p][t]) for p in plants for t in periods)

def total_ad_spend():
    return sum(ad_cost_per_unit[i] * lp.value(ads[i][t]) for i in products for t in [1, 2, 3, 4])

def total_flugel_extra_sales():
    return sum(lp.value(extra_sales["Flugels"][t]) for t in [2, 3, 4, 5])

def total_storage_used(plant):
    return sum(lp.value(inv[plant][i][t]) for i in products for t in [1, 2, 3, 4])

def period1_storage_used(plant):
    return sum(lp.value(inv[plant][i][1]) for i in products)

def total_prod_by_plant(plant):
    return sum(lp.value(prod[plant][i][t]) for i in products for t in periods)

# Run storage sensitivity
storage_results = []

for s in storage_scenarios:
    new_cap_A = BASE_STORAGE_A * s["A_mult"]
    new_cap_B = BASE_STORAGE_B * s["B_mult"]

    # Update RHS of storage cap constraints
    for t in [1, 2, 3, 4]:
        model.constraints[f"StorageCap_A_t{t}"].changeRHS(new_cap_A)
        model.constraints[f"StorageCap_B_t{t}"].changeRHS(new_cap_B)

    # Solve
    solver = lp.PULP_CBC_CMD(msg=False)
    model.solve(solver)

    storage_results.append({
        "Scenario": s["Scenario"],
        "Status": lp.LpStatus[model.status],
        "Plant A Storage Cap": new_cap_A,
        "Plant B Storage Cap": new_cap_B,
        "Profit": lp.value(model.objective),
        "Total Ad Spend": total_ad_spend(),
        "Flugel Extra Sales": total_flugel_extra_sales(),
        "Total Overtime": total_overtime(),
        "Plant A Period 1 Storage Used": period1_storage_used("A"),
        "Plant B Period 1 Storage Used": period1_storage_used("B"),
        "Plant A Total Storage Used": total_storage_used("A"),
        "Plant B Total Storage Used": total_storage_used("B"),
        "Plant A Production": total_prod_by_plant("A"),
        "Plant B Production": total_prod_by_plant("B"),
    })

storage_results_df = pd.DataFrame(storage_results)

baseline_profit = storage_results_df.loc[
    storage_results_df["Scenario"] == "Baseline", "Profit"
].iloc[0]

storage_results_df["Profit Change vs Baseline"] = storage_results_df["Profit"] - baseline_profit

storage_results_df = storage_results_df.round(2)
print(storage_results_df)
```
